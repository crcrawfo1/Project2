```python
# import packages for alpaca, panadas and OS/dotenv to read .env file with api keys
import os
import pandas as pd
import numpy as np
import alpaca_trade_api as tradeapi
# Load .env enviroment variables
from dotenv import load_dotenv
load_dotenv()
import pathlib as Path
```

```python
#helpful maybe?
from datetime import datetime, timedelta

days = 90
startdate = datetime.now() - timedelta(days=days)

# for fixed dates
start = pd.to_datetime('2002-01-01')
end = pd.to_datetime('2020-01-01')
```

```python
# storing alpaca key and secret key in variables 
alpaca_api_key = os.getenv('ALPACA_API_KEY')
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY')

api = tradeapi.REST(alpaca_api_key,alpaca_secret_key,api_version='V2')
```

```python
# set time frame
timeframe = '1D'

# start and 'end' date for covid stock market 


start_date_covid = pd.Timestamp('2019-12-01',tz='America/New_York').isoformat()
end_date_covid = pd.Timestamp('2020-11-20',tz='America/New_York').isoformat()

# start and end date for 2008 recession 
start_date_2008 = pd.Timestamp('2007-07-01',tz='America/New_York').isoformat()
end_date_2008 = pd.Timestamp('2009-07-01',tz='America/New_York').isoformat()

# set tickers (per industry)
tech_tickers = ['AMD', 'AAPL', 'MSFT', 'GOOG', 'INTC']

# get stock data from covid
covid_tech_ticker = api.get_barset(tech_tickers,timeframe,start=start_date_covid,end = end_date_covid).df

# get stock data from 2008 recession
tech_ticker2008 = api.get_barset(tech_tickers,timeframe,start=start_date_2008,end = end_date_2008).df
```

```python
covid_tech_ticker.tail()
```

```python
# create new covid and 2008 dataframe to store ckosing prices of each stock

covid_tech_closingprices = pd.DataFrame()

# get closing prices for all tickers
for ticker in tech_tickers:
    covid_tech_closingprices[ticker]=covid_tech_ticker[ticker]['close']
    
# drop time component on index
covid_tech_closingprices.index = covid_tech_closingprices.index.date
    
covid_tech_closingprices.head()
```

```python
tech_closingprices2008 = pd.DataFrame()

# get closing prices for all tickers
for ticker in tech_tickers:
    tech_closingprices2008[ticker]=tech_ticker2008[ticker]['close']
    
# drop time component on index
tech_closingprices2008.index = tech_closingprices2008.index.date
    
tech_closingprices2008.head()
```

```python
# calculate daily returns for each timeframe
tech_covid_dr = covid_tech_closingprices.pct_change().dropna()
tech_returns2008 = tech_closingprices2008.pct_change().dropna()
```

```python
tech_covid_dr.plot(figsize=(15,10))
```

```python
tech_returns2008.plot(figsize=(20,10))
```

```python
#Name of stocks in portfolio
print('Tech Portfolio with Even Weights for COVID:')
print(covid_tech_closingprices.columns)

tech_covid_weights = np.full((1, len(covid_tech_closingprices.columns)), 1/len(covid_tech_closingprices.columns))[0]
print("Weights")
print(tech_covid_weights)

#expect returns
print('Expected returns')
tech_covid_exp_return = np.sum((tech_covid_dr.mean()*tech_covid_weights)*252)
print(tech_covid_exp_return)

#expected STD/volatility
print('Expected volatility')
tech_covid_exp_volat = np.sqrt(np.dot(tech_covid_weights.T, np.dot(tech_covid_dr.cov()*252, tech_covid_weights)))
print(tech_covid_exp_volat)

#Shape ratio
print('Shape ratio')
tech_covid_sharpe = tech_covid_exp_return/tech_covid_exp_volat
print(tech_covid_sharpe)
```

```python
# running 10k simulations on portfolio
number_ports = 100000
tech_covid_all_weights = np.zeros((number_ports, len(covid_tech_closingprices.columns)))
tech_covid_returns_array = np.zeros(number_ports)
tech_covid_volatility_array = np.zeros(number_ports)
tech_covid_sharpe_array = np.zeros(number_ports)

for index in range(number_ports):
    #weights
    tech_covid_weights = np.array(np.random.random(5))
    tech_covid_weights = tech_covid_weights/np.sum(tech_covid_weights)

    #save weights
    tech_covid_all_weights[index,:] = tech_covid_weights

    #expected returns
    tech_covid_returns_array[index] = np.sum((tech_covid_dr.mean()*tech_covid_weights)*252)

    #expected volatiltiy
    tech_covid_volatility_array[index] = np.sqrt(np.dot(tech_covid_weights.T, np.dot(tech_covid_dr.cov()*252, tech_covid_weights)))

    #sharpe ratio
    tech_covid_sharpe_array[index] = tech_covid_returns_array[index]/tech_covid_volatility_array[index]
```

```python
tech_covid_sharpe_array.max()
```

```python
tech_covid_index_max_shape = tech_covid_sharpe_array.argmax()
tech_covid_index_max_shape
```

```python
tech_covid_weights_array = tech_covid_all_weights[tech_covid_index_max_shape,:]
tech_covid_weights_array
```
